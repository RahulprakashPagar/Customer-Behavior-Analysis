```python
# Missing values to be replaced with median review rating
```

```python
df['Review Rating'] =df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create new column AGE Group
```

```python
labels = ['Young Adult','Adult','middle-aged','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#Create Purchase frquency days
```

```python
Frequency_mapping = {
    'Fortnightly':15,
    'Monthly':30,
    'Quaterly':90,
    'Bi-weekly':14,
    'Weekly':7,
    'Annually':365,
    'Every 3 Months':90
}
df['frequency_of_purchases'] = df['frequency_of_purchases'].map(Frequency_mapping)
df['frequency_of_purchases']
```
